```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv(r'E:\Info_Srping\FWI_Prediction\Data\processed\Interaction_data.csv')
data
```

```python
train = data.iloc[:20000]       
train
```

```python
X = train.drop('fire_weather_index_transformed', axis = 1)
y = train['fire_weather_index_transformed']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
# Save the model
with open(r'E:\Info_Srping\FWI_Prediction\models\Pkl_files\linear_regression_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
# Predict on full training data
y_train_pred = model.predict(X)

train_r2 = r2_score(y, y_train_pred)
train_mse = mean_squared_error(y, y_train_pred)

print('R²:', train_r2)
print('MSE:', train_mse)
```

```python
coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.coef_})
print(coefficients)
```

```python
sns.scatterplot(x=y, y=y_train_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Scatterplot')
plt.savefig('Act_vs_pred.png')
plt.show()
```

```python
import statsmodels.api as sm


X = train.drop('fire_weather_index_transformed', axis=1)
y = train['fire_weather_index_transformed']

# Add constant (intercept) term to X
X_const = sm.add_constant(X)

# Fit OLS model
model_sm = sm.OLS(y, X_const).fit()

# Print detailed regression results summary
print(model_sm.summary())
```
