```python
data = df.drop(columns=['Unnamed: 0','precipitation', 'dew_point_temperature', 'snow_on_ground'])
data
```

```python
import pandas as pd
import numpy as np

# Select numeric variables (exclude target if desired)
vars_for_network = [
    'temperature','relative_humidity','wind_speed','rain_days',
    'fine_fuel_moisture_code','duff_moisture_code','drought_code',
    'build_up_index','initial_spread_index','fire_weather_index'
]

df = data[vars_for_network].copy()

# Choose correlation type: 'pearson' or 'spearman'
corr = df.corr(method='pearson')  # pd.DataFrame (features x features)
```

```python
import networkx as nx

# Parameters
threshold = 0.3        # keep |r| >= 0.30 (tune based on density)
use_abs = False        # set True to ignore sign for filtering

# Create graph
G = nx.Graph()

# Add nodes with degree placeholders (for later styling)
for col in corr.columns:
    G.add_node(col)

# Add edges based on threshold
for i, u in enumerate(corr.columns):
    for j, v in enumerate(corr.columns):
        if j <= i: 
            continue
        r = corr.iloc[i, j]
        if (abs(r) if use_abs else max(r, -r)) >= threshold if use_abs else abs(r) >= threshold:
            G.add_edge(u, v, weight=abs(r), corr=r, sign='pos' if r >= 0 else 'neg')
```

```python
# Compute node sizes based on weighted degree
deg = dict(G.degree(weight='weight'))
node_sizes = [300 + 1200 * (deg[n] / max(deg.values()) if max(deg.values())>0 else 0) for n in G.nodes()]

# Edge attributes for styling
edges = G.edges(data=True)
edge_colors = ['#2ca02c' if d['corr'] >= 0 else '#d62728' for (_,_,d) in edges]
edge_widths = [1 + 4 * d['weight'] for (_,_,d) in edges]

# Layout (spring is good baseline; try kamada_kawai for correlation graphs)
pos = nx.spring_layout(G, seed=42, k=0.6)

plt.figure(figsize=(12, 9))
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='#1f77b4', alpha=0.85)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='white')
nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color=edge_colors, alpha=0.7)

plt.title('Correlation Network (threshold = {:.2f})'.format(threshold))
plt.axis('off')
plt.savefig('Correlation_Network.png')
plt.show()
```

```python
import scipy.stats as stats
import numpy as np

for param in parameters:
    groups = [group[param].values for name, group in data.groupby('FWI_bin')]

    # Normality check using Kolmogorov-Smirnov test against normal distribution N(mean, std)
    normality_ks = []
    for g in groups:
        if len(g) >= 3:
            mean, std = np.mean(g), np.std(g)
            # KS test between sample and normal CDF with sample's mean/std
            ks_stat, ks_p = stats.kstest(g, 'norm', args=(mean, std))
            normality_ks.append(ks_p > 0.05)
    normal_ks = all(normality_ks)

    # Normality check using Anderson-Darling test (returns statistic and critical values)
    normality_ad = []
    for g in groups:
        if len(g) >= 3:
            ad_result = stats.anderson(g, dist='norm')
            # Check if AD statistic is less than critical value at 5% significance level (index 2)
            normality_ad.append(ad_result.statistic < ad_result.critical_values[2])
    normal_ad = all(normality_ad)

    # Equality of variances - Levene's test as before
    if len(groups) > 1:
        variance_homogeneity = stats.levene(*groups)[1] > 0.05
    else:
        variance_homogeneity = True

    print(f"{param}: KS Normal? {normal_ks}, AD Normal? {normal_ad}, Equal variances? {variance_homogeneity}")
```
