```python
import pandas as pd
import pickle
from sklearn.preprocessing import PowerTransformer, StandardScaler , OneHotEncoder
```

```python
data = pd.read_csv(r"E:\Info_Srping\FWI_Prediction\Data\processed\filtered_data_NN.csv")
data
```

```python
data = data.drop(['Unnamed: 0','wind_direction','snow_on_ground','daily_severity_rating','latitude','longitude', 'elevation','visibility'], axis = 1)
data
```

```python
non_nd_col = ['fine_fuel_moisture_code','duff_moisture_code', 'drought_code','build_up_index', 'initial_spread_index','fire_weather_index','wind_speed']
```

```python
# Dictionary to hold fitted power transformers per column
power_transformers = {}

# Fit power transformer per specified column and save each transformer object
for col in non_nd_col:
    pt = PowerTransformer(method='yeo-johnson', standardize=False)
    transformed = pt.fit_transform(data[col].values.reshape(-1, 1))
    data[col + '_transformed'] = transformed.flatten()
    power_transformers[col] = pt  # Save fitted transformer

# Save the power transformers dictionary as a pkl
with open(r'E:\Info_Srping\FWI_Prediction\models\Pkl_files\power_transformers.pkl', 'wb') as f:
    pickle.dump(power_transformers, f)
```

```python
# Fit StandardScaler on all numerical columns (including newly added transformed columns)
scaler = StandardScaler()

# Select numeric columns from data
numeric_cols = data.select_dtypes(include=['float64', 'int']).columns

# Fit and transform scaler on all numeric features together
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

# Save the fitted scaler as a pkl
with open(r'E:\Info_Srping\FWI_Prediction\models\Pkl_files\standard_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore', dtype=int)
ohe.fit(data[['country']])

# Save the fitted encoder to a pickle file
with open(r'E:\Info_Srping\FWI_Prediction\models\Pkl_files\one_hot_encoder.pkl', 'wb') as f:
    pickle.dump(ohe, f)
```
