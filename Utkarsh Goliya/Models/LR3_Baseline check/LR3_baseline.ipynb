```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv(r'E:\Info_Srping\FWI_Prediction\Data\processed\Interaction_data.csv')
data
```

```python
train = data.iloc[:20000]       
train
```

```python
train.columns
```

```python
features_sub = ['fine_fuel_moisture_code_transformed',
                'duff_moisture_code_transformed',
               'drought_code_transformed',
                'build_up_index_transformed',
               'initial_spread_index_transformed']

features_full = ['fine_fuel_moisture_code_transformed',
                'duff_moisture_code_transformed',
               'drought_code_transformed',
                'build_up_index_transformed',
               'initial_spread_index_transformed','temperature', 
                'dew_point_temperature', 'relative_humidity',
                'atmospheric_pressure', 'precipitation', 
                 'rain_days']
```

```python
# Split into features and target
X_sub = train[features_sub]
X_full = train[features_full]
y = train['fire_weather_index_transformed']
```

```python
# Train-test split
X_train_sub, X_test_sub, y_train, y_test = train_test_split(X_sub, y, test_size=0.2, random_state=42)
X_train_full, X_test_full, _, _ = train_test_split(X_full, y, test_size=0.2, random_state=42)
```

```python
# Linear regression models
lr_sub = LinearRegression().fit(X_train_sub, y_train)
lr_full = LinearRegression().fit(X_train_full, y_train)
```

```python
# Evaluate
print("Sub-indices R²:", r2_score(y_test, lr_sub.predict(X_test_sub)))
print("Full model R² :", r2_score(y_test, lr_full.predict(X_test_full)))
print("Sub-indices MSE:", mean_squared_error(y_test, lr_sub.predict(X_test_sub)))
print("Full model MSE :", mean_squared_error(y_test, lr_full.predict(X_test_full)))
```
