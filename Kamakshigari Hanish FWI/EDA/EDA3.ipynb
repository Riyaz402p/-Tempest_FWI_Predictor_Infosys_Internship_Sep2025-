```python
import pandas as pd

elev_bins = [0, 500, 1000, 1500, 2000, 2500]
elev_labels = ['0–500', '500–1000', '1000–1500', '1500–2000', '2000–2500']

# Create a new column for elevation band
df_sample['elevation_band'] = pd.cut(df_sample['elevation'], bins=elev_bins, labels=elev_labels, right=False)

# Group by elevation band and calculate summary statistics
table = df_sample.groupby('elevation_band')['fire_weather_index'].agg(
    avg_fwi='mean',
    median_fwi='median',
    count='size'
).reset_index()

print(table)
```

```python
predictors = ['temperature', 'rain_days', 'relative_humidity', 'wind_speed']

for var in predictors:
    sns.scatterplot(x=var, y='fine_fuel_moisture_code', data=df_sample)
    plt.title(f'FFMC vs {var.capitalize()}')
    plt.xlabel(var.capitalize())
    plt.ylabel('FFMC')
    plt.savefig(f'{var}.png')
    plt.show()
```

```python
# FFMC vs Rain_days with color by Temperature
sns.scatterplot(data=df_sample, x='rain_days', y='fine_fuel_moisture_code', hue='temperature', palette='coolwarm', alpha=0.7)
plt.title('FFMC vs Rain Days colored by Temperature')
plt.xlabel('Rain Days')
plt.ylabel('FFMC')
plt.savefig('FFMC_vs_Rain.png')
plt.show()
```

```python
# FFMC vs Wind_speed with color by Relative Humidity
sns.scatterplot(data=df_sample, x='wind_speed', y='fine_fuel_moisture_code', hue='relative_humidity', palette='viridis', alpha=0.7)
plt.title('FFMC vs Wind Speed colored by Relative Humidity')
plt.xlabel('Wind Speed')
plt.ylabel('FFMC')
plt.savefig('FFMC_vs_Wind.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

sc = plt.scatter(df_sample['rain_days'], df_sample['fine_fuel_moisture_code'], c=df_sample['temperature'], cmap='coolwarm', alpha=0.7)
plt.title('FFMC vs Rain Days colored by Temperature')
plt.xlabel('Rain Days')
plt.ylabel('FFMC')
plt.colorbar(sc, label='Temperature')
plt.show()

sc2 = plt.scatter(df_sample['wind_speed'], df_sample['fine_fuel_moisture_code'], c=df_sample['relative_humidity'], cmap='viridis', alpha=0.7)
plt.title('FFMC vs Wind Speed colored by Relative Humidity')
plt.xlabel('Wind Speed')
plt.ylabel('FFMC')
plt.colorbar(sc2, label='Relative Humidity')
plt.show()
```

```python
sns.scatterplot(data=df_sample, x='relative_humidity', y='fine_fuel_moisture_code', hue='temperature', palette='viridis', alpha=0.7)
plt.title('FFMC vs temperature colored by Relative Humidity')
plt.xlabel('relative_humidity')
plt.ylabel('FFMC')
plt.savefig('FFMC_vs_RH.png')
plt.show()
```
